## ‚úèÔ∏è Relembrando: o que √© um agente reflexivo simples?

Um **agente reflexivo**:
- N√£o tem mem√≥ria ou modelo interno.
- Age **apenas** em resposta √† percep√ß√£o atual.
- Usa **regras condicionais simples**: se perceber X, faz Y.

---

## üíª Implementa√ß√£o b√°sica em Python

Vamos criar um agente que se move por um ambiente simples:  
- Se perceber um **obst√°culo**, desvia.  
- Se o caminho est√° **livre**, anda para frente.

```python



````python
class AgenteReflexivoSimples:
    def __init__(self):
        pass  # sem mem√≥ria ou estado interno

    def agir(self, percepcao):
        if percepcao == "obstaculo":
            return "desviar"
        else:
            return "andar"
````

## üîç Testando o agente com entradas simples

````python
# Instanciando o agente
agente = AgenteReflexivoSimples()

# Lista de percep√ß√µes simuladas
percepcoes = ["livre", "livre", "obstaculo", "livre", "obstaculo", "obstaculo"]

# A√ß√µes tomadas pelo agente
for i, percepcao in enumerate(percepcoes):
    acao = agente.agir(percepcao)
    print(f"Passo {i+1}: percep√ß√£o={percepcao} => a√ß√£o={acao}")
````

## üî¨ Visualiza√ß√£o simples do comportamento
Podemos plotar um gr√°fico para ver as percep√ß√µes e a√ß√µes ao longo do tempo.

````python
import matplotlib.pyplot as plt

acoes = [agente.agir(p) for p in percepcoes]

plt.figure(figsize=(10,4))
plt.plot(range(len(percepcoes)), percepcoes, label="Percep√ß√£o", marker='o')
plt.plot(range(len(acoes)), acoes, label="A√ß√£o", marker='x')
plt.xlabel("Passo")
plt.ylabel("Estado")
plt.title("Comportamento do Agente Reflexivo Simples")
plt.legend()
plt.grid(True)
plt.show()
````
